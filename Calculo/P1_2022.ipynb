Sabemos que para a função ser contínua, os dois limites laterais tem que ser o mesmo, logo iremos calcular eles para obter um sistema de equações para descobrirmos o valor de a.

```python
import sympy as sp
x = sp.symbols('x')
a = sp.symbols('a')
sp.limit(x**2+a*x,x,1,'-')
```

```python
import sympy as sp
x = sp.symbols('x')
a = sp.symbols('a')
sp.limit(a**2*x**2-1,x,1,'+')
```

Para conferir se a resposta está certa, vamos plotar os 2 gráficos e ver os pontos que eles se interceptam.

Para saber se existem assíntotas horizontais, analisaremos o limite ±∞: 

```python
import sympy as sp
x = sp.symbols('x')
sp.limit((2*x**2+x-3)/(2*x**2-x),x,sp.oo)
```

```python
import sympy as sp
x = sp.symbols('x')
sp.limit((2*x**2+x-3)/(2*x**2-x),x,-sp.oo)
```

Portanto 1 é assíntota horizontal para o infinito negativo e para o positivo

Para ter assíntota vertical em x=a, algum dos limites laterais de f em a precisa ser infinito ou menos infinito. Bons valores para conferirmos são os que zeram o denominador de f que são:

```python
import sympy as sp
x = sp.symbols('x')
sp.solve(2*x**2-x)
```

0 e 1/2 são os valores que zeram o denominador, portanto vamos analisar esses pontos separadamente.
