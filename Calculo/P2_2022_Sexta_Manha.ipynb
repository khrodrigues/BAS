Agora vamos derivar a equação inicial ($x^2+xy+y^2=7$) implicitamente com relação a x para obtermos o coeficiente angular das retas. Não podemos fazer isso com python, por isso teremos que recorrer aos métodos manuais dessa vez.
$$2x+y+x*y'+2y*y'=0$$
Isolando $y'$ teremos:
$$y'=\frac{-2x-y}{x+2y}$$
Aplicando agora os pontos em $y'$:

```python
import sympy as sp
def f(x,y):
    return (-2*x-y)/(x+2*y)
x = sp.symbols('x', real=True)
y = sp.symbols('y', real=True)
f(7**(1/2),0)
```

```python
import sympy as sp
def f(x,y):
    return (-2*x-y)/(x+2*y)
x = sp.symbols('x', real=True)
y = sp.symbols('y', real=True)
f(-7**(1/2),0)
```

Como $s=r*\theta$ temos:
$$\frac{ds}{dt}=r\frac{d\theta}{dt}$$
Sabendo que $\frac{ds}{dt}=7 m/s$ isolando $\frac{d\theta}{dt}$ teremos que $$\frac{d\theta}{dt}=\frac{7}{r}=\frac{7}{100}$$

Sabendo que a distancia entre eles é de 200, vamos agora saber o ângulo $\theta$:
$$z^2=100^2*(5-4*Cos(\theta))$$
$$200^2=100^2*(5-4*Cos(\theta))$$
Para isso, contaremos com o auxílio do Python e da biblioteca SymPy:

```python
import sympy as sp
x, y, alpha = sp.symbols('x y alpha')
sp.solve(200**2-100**2*(5-4*sp.cos(alpha)))
```
