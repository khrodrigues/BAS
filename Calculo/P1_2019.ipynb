Agora vamos resolver o sistema. 

```python
import sympy as sp
x, y, z, a, b= sp.symbols('x y z a b')

eq1=sp.Eq(x+a*y,2)
eq2=sp.Eq((a+1)*x+2*y+(a+2)*z,3*b-2)
eq3=sp.Eq(x+(a)*y+(a+2)*z,(b+2))

sp.linsolve((eq1,eq2,eq3),(x,y,z))
```

Com essa solução geral, vemos por que o sistema não tem solução para A=1 e A=-2, então essa é a solução para valores de A diferente de -2 e 1

Agora usando A=1, teremos as seguintes soluções:

```python
import sympy as sp
x, y, z, a, b= sp.symbols('x y z a b')

a=1
eq1=sp.Eq(x+a*y,2)
eq2=sp.Eq((a+1)*x+2*y+(a+2)*z,3*b-2)
eq3=sp.Eq(x+(a)*y+(a+2)*z,(b+2))

sp.linsolve((eq1,eq2,eq3),(x,y,z))
```

Como podemos ver, não tem solução para A=1.

Agora usando A=-2, teremos as seguintes soluções:

```python
import sympy as sp
x, y, z, a, b= sp.symbols('x y z a b')
a=-2
eq1=sp.Eq(x+a*y,2)
eq2=sp.Eq((a+1)*x+2*y+(a+2)*z,3*b-2)
eq3=sp.Eq(x+(a)*y+(a+2)*z,(b+2))

sp.linsolve((eq1,eq2,eq3),(x,y,z))
```

Para a matriz ser inversível, sabemos que a determinante dela deve ser diferente de 0, logo temos que calcular a sua deteterminante e faremos da mesma forma que na questão anterior.

```python
import sympy as sp
k = sp.symbols('k')
Matriz=sp.Matrix([[1,-2,0],[3*k,8+2*k,k-1],[0,8*k+8,0]])
Matriz.det()
```

Igualaremos a 0 e resolveremos essa equação para achar os valores de k para os quais a matriz não é invertível.

```python
import sympy as sp
sp.solve(8-8*k**2)
```
