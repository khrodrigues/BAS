$$a_1*A+a_2*A+...+a_m*A^m=-a_0*Id$$
$$-\frac{a_1}{a_0}*A-\frac{a_2}{a_0}*A^2-...-\frac{a_m}{a_0}*A^m=Id$$
$$A\left \{-\frac{a_1}{a_0}*A^{-1}-\frac{a_2}{a_0}*A-...-\frac{a_m}{a_0}*A^{m-1}  \right \}=Id$$
Logo a inversa é: $A^{-1}=-\frac{a_1}{a_0}*A^{-1}-\frac{a_2}{a_0}*A-...-\frac{a_m}{a_0}*A^{m-1}$
